```python
# Import necessary module
import json

# Function to process the gene symbol
def extract_from_json(json_input):
    # Create a list to save the results
    gene_list = []

    # Read in json data
    with open(json_input, "r", encoding = "utf-8") as f:
        datas = json.load(f)
        for data in datas["associations"]:
            dict_1 = data["gene"]
            gene_symbol = dict_1["symbol"]
            gene_list.append(gene_symbol)
    
    return gene_list
```

### Main Process

```python
if __name__ == "__main__":
    json_input = [r"/home/terry-liu/桌面/xgboost/data/wnt_pathway.json",
                  r"/home/terry-liu/桌面/xgboost/data/TGF_beta.json",
                  r"/home/terry-liu/桌面/xgboost/data/Ras.json"]
    print(f" [i] For conversion to Ensembl_IDs, please visit: https://biit.cs.ut.ee/gprofiler/convert")
    for file in json_input:
        gene_list = extract_from_json(file)
        print("\n".join(gene_list))
    
```
